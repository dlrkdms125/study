```python
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
import gdown
```

```python
# Google Drive 파일 다운로드 함수
def download_from_drive(drive_url, output_path):
    file_id = drive_url.split('/d/')[1].split('/')[0]
    download_url = f"https://drive.google.com/uc?id={file_id}"
    gdown.download(download_url, output_path, quiet=False)
```

```python
# MediaPipe Pose로 신체 랜드마크 추출
def get_pose_landmarks(image_path):
    mp_pose = mp.solutions.pose
    pose = mp_pose.Pose(static_image_mode=True)

    # 이미지 로드 및 RGB 변환
    image = cv2.imread(image_path)
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # 포즈 추정 수행
    results = pose.process(rgb_image)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark
        return landmarks, image  # 랜드마크와 원본 이미지 반환
    else:
        print("No pose landmarks detected.")
        return None, image
```

```python
# 특정 랜드마크 간 거리 계산 함수
def calculate_distance(landmark1, landmark2, image_shape):
    x1, y1 = int(landmark1.x * image_shape[1]), int(landmark1.y * image_shape[0])
    x2, y2 = int(landmark2.x * image_shape[1]), int(landmark2.y * image_shape[0])
    distance = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance
```

```python
# 신체 부위 측정 및 위치 반환 함수 (어깨, 허리, 팔뚝 등)
def measure_body_parts(landmarks, image_shape):
    body_parts = {}

    # 어깨 너비 (왼쪽 어깨 - 오른쪽 어깨)
    body_parts["shoulder_width"] = {
        "value": calculate_distance(
            landmarks[mp.solutions.pose.PoseLandmark.LEFT_SHOULDER],
            landmarks[mp.solutions.pose.PoseLandmark.RIGHT_SHOULDER],
            image_shape
        ),
        "position": (
            int((landmarks[mp.solutions.pose.PoseLandmark.LEFT_SHOULDER].x +
                 landmarks[mp.solutions.pose.PoseLandmark.RIGHT_SHOULDER].x) / 2 * image_shape[1]),
            int((landmarks[mp.solutions.pose.PoseLandmark.LEFT_SHOULDER].y +
                 landmarks[mp.solutions.pose.PoseLandmark.RIGHT_SHOULDER].y) / 2 * image_shape[0]) - 20
        )
    }

    # 허리 너비 (왼쪽 엉덩이 - 오른쪽 엉덩이)
    body_parts["waist_width"] = {
        "value": calculate_distance(
            landmarks[mp.solutions.pose.PoseLandmark.LEFT_HIP],
            landmarks[mp.solutions.pose.PoseLandmark.RIGHT_HIP],
            image_shape
        ),
        "position": (
            int((landmarks[mp.solutions.pose.PoseLandmark.LEFT_HIP].x +
                 landmarks[mp.solutions.pose.PoseLandmark.RIGHT_HIP].x) / 2 * image_shape[1]),
            int((landmarks[mp.solutions.pose.PoseLandmark.LEFT_HIP].y +
                 landmarks[mp.solutions.pose.PoseLandmark.RIGHT_HIP].y) / 2 * image_shape[0]) - 20
        )
    }

    # 왼팔 길이 (왼쪽 어깨 - 왼쪽 팔꿈치 - 왼쪽 손목)
    left_arm_length = calculate_distance(
        landmarks[mp.solutions.pose.PoseLandmark.LEFT_SHOULDER],
        landmarks[mp.solutions.pose.PoseLandmark.LEFT_ELBOW],
        image_shape
    ) + calculate_distance(
        landmarks[mp.solutions.pose.PoseLandmark.LEFT_ELBOW],
        landmarks[mp.solutions.pose.PoseLandmark.LEFT_WRIST],
        image_shape
    )
    body_parts["left_arm_length"] = {
        "value": left_arm_length,
        "position": (
            int(landmarks[mp.solutions.pose.PoseLandmark.LEFT_ELBOW].x * image_shape[1]),
            int(landmarks[mp.solutions.pose.PoseLandmark.LEFT_ELBOW].y * image_shape[0]) - 20
        )
    }

    # 오른팔 길이 (오른쪽 어깨 - 오른쪽 팔꿈치 - 오른쪽 손목)
    right_arm_length = calculate_distance(
        landmarks[mp.solutions.pose.PoseLandmark.RIGHT_SHOULDER],
        landmarks[mp.solutions.pose.PoseLandmark.RIGHT_ELBOW],
        image_shape
    ) + calculate_distance(
        landmarks[mp.solutions.pose.PoseLandmark.RIGHT_ELBOW],
        landmarks[mp.solutions.pose.PoseLandmark.RIGHT_WRIST],
        image_shape
    )
    body_parts["right_arm_length"] = {
        "value": right_arm_length,
        "position": (
            int(landmarks[mp.solutions.pose.PoseLandmark.RIGHT_ELBOW].x * image_shape[1]),
            int(landmarks[mp.solutions.pose.PoseLandmark.RIGHT_ELBOW].y * image_shape[0]) - 20
        )
    }

    return body_parts

```

```python
# 두 이미지의 신체 부위 비교 및 결과 표시 함수
def annotate_comparison(image1, measurements1, image2, measurements2):
    annotated_image = image2.copy()

    for part in measurements1.keys():
        diff = measurements2[part]["value"] - measurements1[part]["value"]

        # 텍스트 위치 설정 (각 부위 위에 표시)
        text_position = measurements2[part]["position"]

        # 차이를 텍스트로 이미지에 표시
        cv2.putText(
            annotated_image,
            f"{part}: {diff:.2f} pixels",
            text_position,
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0) if diff > 0 else (0, 0, 255),  # 증가: 초록색 / 감소: 빨간색
            2
        )

    return annotated_image
```

```python
# Google Drive 이미지 경로 설정
image1_url = "https://drive.google.com/file/d/1g2qRSEoKcGWVG5DgvjOU51twwGKwt9d0/view?usp=drive_link"
image2_url = "https://drive.google.com/file/d/1aaBIkxTQvZ1ufgICJQOrfmxk5DKQXGIm/view?usp=drive_link"
image1_path = "image1.jpg"
image2_path = "image2.jpg"
```

```python
# Google Drive에서 이미지 다운로드
download_from_drive(image1_url, image1_path)
download_from_drive(image2_url, image2_path)
```

```python
# 두 이미지에서 랜드마크 추출 및 신체 부위 측정
landmarks1, img1 = get_pose_landmarks(image1_path)
landmarks2, img2 = get_pose_landmarks(image2_path)

if landmarks1 and landmarks2:
    measurements1 = measure_body_parts(landmarks1, img1.shape)
    measurements2 = measure_body_parts(landmarks2, img2.shape)

    # 변화량을 각 부위 위에 표시한 이미지 생성
    annotated_result = annotate_comparison(img1, measurements1, img2, measurements2)

    # 결과 이미지 저장 및 표시
    cv2.imwrite("comparison_result.jpg", annotated_result)

    from matplotlib import pyplot as plt

    plt.figure(figsize=(10, 5))
    plt.title("Comparison Result")
    plt.imshow(cv2.cvtColor(annotated_result, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
